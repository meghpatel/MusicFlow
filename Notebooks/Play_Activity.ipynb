```python
# Get sparsity of the data

# Create empty dictionary 
from pandas import isna


frequency = {}
for col in list(columns):
    frequency[col] = 0

for i, row in df.iterrows():
    for col in list(columns):
        if pd.isna(row[col]):
            frequency[col] += 1
```

```python
len(frequency.keys())
```

```python
total_rows = df.shape[0]
```

```python
for key, val in frequency.items():
    frequency[key] = (val/total_rows) * 100
```

```python
# 100% Sparsity
for key, val in frequency.items():
    if val == 100:
        print (key, val)
```

```python
# 99%
for key, val in frequency.items():
    if val >= 99 and val < 100:
        print (key, val)
```
